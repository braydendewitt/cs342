```python
import torch
from PIL import Image
import numpy as np
```

```python
torch.zeros(10)
# Order one tensor with 10 value zeros
```

```python
torch.ones(10)
# Order one tensor with 10 value ones
```

```python
torch.ones([4,5])
# Order two tensor of 4x5 value ones
```

```python
v = torch.ones(5)
print( v.dtype, v.shape )
# Data type and shape
```

```python
v = torch.arange(100)
v
# Order one tensor of 100 values
```

```python
v.shape
```

```python
m = v.view((10,10))
m
# Changes the shape of the tensor to be 10x10
# v.view(-1) flattens the tensor
```

```python
m.shape
# Now check that shape is 10x10
```

```python
I = Image.open('cat.jpg')
I
# displays image
```

```python
np.array(I)
# get pixel values of the image
```

```python
np.array(I).shape
# shape of numpy array (240x320x3) (height, width, 3 color channels RGB)
```

```python
from torchvision import transforms
image_to_tensor = transforms.ToTensor()
image_tensor = image_to_tensor(I)
image_tensor
# Convert image to a tensor, must import transforms from torchvision
```

```python
image_tensor.shape
# Now get tensor shape
```

```python
tensor_to_image = transforms.ToPILImage()
tensor_to_image(image_tensor)
# Transform tensor to image (using .ToPILImage)
```
